證交所資料爬取

```python
import datetime
import pandas as pd
import time


startdate = datetime.datetime(2022,9,1)
enddate = datetime.datetime(2022,9,5)
days = (enddate-startdate).days+1

for i in range(days):
    try:
        time.sleep(5)
        dates = (startdate + datetime.timedelta(days = i)).date()
        date = str(dates).replace("-","")
        df = pd.read_html("https://www.twse.com.tw/exchangeReport/MI_INDEX?response=html&date="+date+"&type=ALLBUT0999")
        df = df[-1]
        df.columns = [ i[2] for i in df.columns]
        df.index = df.證券代號
        df = df.drop(columns = '證券代號')
        df.to_csv(date+'.csv')
        
    except:
        continue
```
